**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
set(properties['borough'])
```

**We look at the set of entries in the borough column and see that we have too many entries. It looks like the entreis that are in all capital letters need to be removed. Further inspection also finds 3 other entries (City of London, England, Inner London) that are not borough names either. We go through each row of the dataframe and look at the borough column. If this is all capital letters or one of the other 3 non-boroughs, we note the index and then remove all indices from the dataframe.

```python
not_a_borough = ['City of London', 'England', 'Inner London']
```

```python
drop_list = []
for i in range(properties.shape[0]):
   #for ele in properties.iloc[i, 0]:
         
       # use isupper() 
    if (properties.iloc[i, 0].isupper()) or properties.iloc[i, 0] in not_a_borough:
           drop_list.append(i)
```

```python
properties = properties.drop(properties.index[drop_list])
```

**Looking at the set of 'borough' labels we now have the expected 32 different boroughs.

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

**We find the set of borough names which we will pass to a function to return the ratio of price increase.

```python
borough_names = set(properties['borough'])
```

```python

def create_price_ratio(borough):
    #create subset of data for the particular borough
    subset = properties[properties['borough'] == borough]
    
    #subset only data from 1998 for a particular borough and compute mean
    avg_price_1998 = subset[(subset['date'] >= '1998-01-01')\
                             & (subset['date'] <= '1998-12-31')]\
                                ['avg_price'].mean()
    
    #subset only data from 2018 for a particular borough and compute mean
    avg_price_2018 = subset[(subset['date'] >= '2018-01-01')\
                             & (subset['date'] <= '2018-12-31')]\
                                ['avg_price'].mean()
    
    #compute ratio
    ratio = avg_price_2018 / avg_price_1998
    
    #return the ratio and borough name
    return (borough, ratio)
    
```

**We will use a dictionary to store each borough name with its associated ratio increase from 1998 to 2018.

```python
ratio_dict = {}
for name in borough_names:
    name, value = create_price_ratio(name)
    ratio_dict[name] = value
    
```

**Now we create a dataframe using the dictionary of borough names and ratios. We then order this according to the ratios and graph of a horizontal bar graph. 

```python
ratio_df = pd.DataFrame(list(ratio_dict.items()), columns=['borough', 'ratio'])
```

```python
ratio_df_sorted = ratio_df.sort_values('ratio', ascending=True)
```

```python
plt.figure(figsize=(10,20))
plt.barh(ratio_df_sorted['borough'], ratio_df_sorted['ratio'])
for index, value in enumerate(ratio_df_sorted['ratio']):
    plt.text(value, index, str(round(value, 2)))
plt.title('Ratio increase of London boroughs from 1998 to 2018.')
plt.xlabel('Ratio increase')
plt.show()
```
